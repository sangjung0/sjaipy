```python
import os

from pathlib import Path
from IPython.display import Audio

DATA_DIR = os.getenv("CONTAINER_DATA_DIR")
```

```python
from sjaipy.datasets.l_hotse import LibriSpeech
```

```python
path = Path(f"{DATA_DIR}/.datasets/libri_speech")
```

```python
libri_speech = LibriSpeech(path)
```

```python
libri_speech.prepare()
```

```python
dev_set = libri_speech.load_dev_clean(task=["asr", "diarization"])
```

```python
idx = 12
id_set = set()
for i, sample in enumerate(dev_set):
    if sample.id in id_set:
        raise ValueError(f"Duplicate recording ID found: {sample.id}")
    id_set.add(sample.id)

    if i == idx:
        break
```
